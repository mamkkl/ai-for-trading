# Example of a PCA Risk Factor Model

This workbook and repo is based off of Udacity's AI for Trading lesson on Factor Models.  

The risk factor model that we will be modeling is defined as:

$$
\textbf{r} = \textbf{B}\textbf{f} + \textbf{S}
$$

where 

- $\textbf{r}$ is a matrix containing the asset returns
- $\textbf{B}$ is a matrix representing the factor exposures
containing
- $\textbf{f}$ is the matrix of factor returns
- $\textbf{S}$ is the idiosyncratic risk (also known as the company specific risk).

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from data_loader import DataLoader
from pca_factor_model import PCAFactorModel
from zipline.pipeline.factors import AverageDollarVolume

%matplotlib inline
```

## Get returns

The stock returns are compiled using Quantopian's Zipline module with data provided from Quotemedia.

```python
start_date = "2011-01-05" #Both dates must be valid trading dates
end_date = "2016-01-05"
bundle_name = 'm4-quiz-eod-quotemedia'
pipeline_screen = AverageDollarVolume(window_length=120).top(500)

dl = DataLoader(pipeline_screen,bundle_name, start_date, end_date, \
                 exchange_calendar="NYSE",
                 data_frequency="daily")

# Get the returns for the start and end date. 
returns = dl.load_data()

# Display the first rows of the returns
returns.head()
```

## Instantiate PCAFactorModel

The PCAFactorModel calculates the individual components of the Risk Factor Model.  The inputs are:

- returns: pandas dataframe
- num_components: the number of PCA 

```python
pca_factor_model = PCAFactorModel(num_components=50, annualization_factor=252)
pca_factor_model.fit(returns)
```

## Factor Betas

Factor Betas, **B**, is a matrix of the coordinates of the original principle components.

```python
pca_factor_model.factor_betas_
```

## Factor Returns

The factor returns, **f**, are calculated from Scikit-Learn's PCA() class.  

The output is a dataframe of returns with the **new** basis. 

```python
pca_factor_model.factor_returns_
```

```python
pca_factor_model.plot_factor_returns(top_n_factors=5)
```

## Residuals

The residuals are the difference between the actual returns and the pca fit returns, common_returns **r_pca**.


$$\textbf{S} = \textbf{r} - \textbf{r_pca}$$ 

where

- **r_pca** is the common_returns = factor_betas * factor_returns

$$ \textbf{r_pca} = \textbf{Bf}$$


```python
pca_factor_model.residuals_
```

```python
pca_factor_model.common_returns_
```

## Idiosyncratic Risk Matrix

The Idiosyncratic Risk Matrix is the covariance matrix of the residuals, **S**. 


The function outputs a panda dataframe of the annualized covariance of residuals in the diagonal with all zeros in the off-diagonal elements. 

```python
pca_factor_model.idiosyncratic_var_matrix_
```

```python
pca_factor_model.idiosyncratic_var_vector_
```

## Factor Covariance Matrix

The factor covariance matrix, $\textbf{F}$, equation is:

$$
\textbf{F} = \frac{1}{N -1}\textbf{f} \textbf{f}^T
$$

where

- $\textbf{f}$ is the matrix of factor returns
- $N$ is the number of elements in $\textbf{f}$

The function outputs a panda dataframe of the annualized covariance in the diagonal with all zeros in the off-diagonal elements. 

```python
pca_factor_model.factor_cov_matrix_
```

## Explained Variance Ratio

The explained variance ratio is the percentage of variance explained by each of the selected components.
